```python
import csv

filelist = [
    r"D:\Onedrive\ACGN相关\切煤代购\夜宵称重表\其他.csv",
    r"D:\Onedrive\ACGN相关\切煤代购\夜宵称重表\包税09.csv",
]



output_path = r"D:\Onedrive\ACGN相关\切煤代购\夜宵称重表"
output_path +=r"\20240807.csv"
```

```python
cn_group = []
gram_group = []
price_group = []


for filepath in filelist:
    with open(filepath, 'r', newline='', encoding='utf-8-sig') as csvfile:
        # 创建CSV阅读器
        csvreader = csv.reader(csvfile)
        for row in csvreader:
            dataGroup = [item for item in row]
            cn = str(dataGroup[0])
            cn = cn.lower()
            cn = (cn.split())[0]
            print(cn)
            gram = round(float(dataGroup[1]), 2)
            price = round(float(dataGroup[2]), 2)
            try:
                # cn存在
                cn_index = cn_group.index(cn)
                gram_group[cn_index] += gram
                price_group[cn_index] += price
            except ValueError:
                # cn不存在
                cn_group.append(cn)
                gram_group.append(gram)
                price_group.append(price)

output_list = [f"{cn},{round(float(gram), 2)},{price}" for cn, gram, price in zip(
    cn_group, gram_group, price_group)]

with open(output_path, 'w', newline='', encoding='utf-8-sig') as csvfile:
    # 创建CSV写入器
    csvwriter = csv.writer(csvfile)

    # 写入标题行
    csvwriter.writerow(["cn", "重量", "肾额"])

    # 将每个元素写入CSV文件的一行
    for row in output_list:
        csvwriter.writerow(row.split(','))
```
